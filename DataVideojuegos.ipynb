```python
import pandas as pd

#Cargamos el dataset desde la ruta proporcionada en un DataFrame llamado df
df = pd.read_csv('C:/Users/Alan/Desktop/9mil.csv', header=0)

"print(df)"
df.head()
```

```python
# Reemplazar los valores "TBD" con NaN en la columna 'Release_Date'
df['Release_Date'] = df['Release_Date'].replace('TBD', pd.NA)

# Eliminar las filas que contienen NaN en la columna 'Release_Date'
df = df.dropna(subset=['Release_Date'])

# Mostrar el DataFrame actualizado sin los valores "TBD" en la columna 'Release_Date'
print(df)
```

```python
# Verificar si hay IDs duplicados en el Dataset
duplicados_ids = df.duplicated(subset=['ID'], keep=False)

# Filtramos y mostramos solo los registros duplicados en funcin de 'ID'
registros_duplicados_por_id = df[duplicados_ids]

# Si existen IDs duplicados, imprimimos
if not registros_duplicados_por_id.empty:
    print('IDs duplicados:')
    print(registros_duplicados_por_id.head())
else:
    print('No existen IDs duplicados en el Dataset.')
    

registros_duplicados_por_id.head()
```

```python
# Verificamos si hay registros vacos en el Dataset
registros_vacios = df.isnull().any(axis=1)

# Filtrar y mostrar solo los registros con valores nulos
registros_con_valores_nulos = df[registros_vacios]

# Si existen registros con valores nulos, imprimimos
if not registros_con_valores_nulos.empty:
    print('Registros con valores nulos:')
    print(registros_con_valores_nulos)
    
    # Eliminamos los registros con valores nulos del DataFrame original
    df = df.dropna()

    # Imprimimos el contenido del DataFrame actualizado
    print('Dataset sin campos vacos:')
    print(df)
else:
    print('No existen registros con valores nulos en el Dataset.')
```

```python
# Mostramos la cantidad de datos en el dataset
num_datos = df.shape[0]
print(f'La cantidad de datos en el dataset es: {num_datos}')
```

```python
# Verificar si hay ttulos de videojuegos duplicados en el Dataset
duplicados_titulos = df.duplicated(subset=['Title'], keep=False)

# Filtrar y mostrar solo los ttulos de videojuegos duplicados
titulos_duplicados = df.loc[duplicados_titulos, 'Title']

# Imprimir los ttulos de videojuegos duplicados
print('Ttulos de videojuegos duplicados:')
print(titulos_duplicados)
```

```python
# Eliminar los ttulos de videojuegos duplicados del DataFrame original
df = df.drop_duplicates(subset=['Title'], keep='first')

# Mostrar el DataFrame actualizado sin ttulos duplicados
print('DataFrame sin ttulos duplicados:')
print(df)
```
